```python
data_svm = pd.read_excel(r'D:\hk\Raisin_Dataset\Raisin_Dataset.xlsx', engine="openpyxl")
```

```python
print("Unique values in data['Class']:", data_svm['Class'].unique())
```

```python
data_svm['Class'] = data_svm['Class'].map({'Kecimen': 1, 'Besni': 0})
```

```python
print("Unique values in data['Class']:", data_svm['Class'].unique())
```

```python
data_svm.head()
```

```python

x_data = data_svm.drop(["Class"], axis = 1)

y_data = data_svm.Class.values
```

```python
print("Unique values in data['Class']:", data_svm['Class'].unique())
```

```python
x_data = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state = 1)
```

```python
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.svm import SVC
svm = SVC(random_state = 1)

svm.fit(x_train, y_train)

print("print accuracy of svm algo:", svm.score(x_test, y_test))
```

```python
test_score_list = []
train_score_list = []

for i in range(1, 200):
    svm2 = SVC(C=i)
    svm2.fit(x_train, y_train)
    train_score_list.append(svm2.score(x_train, y_train))
    test_score_list.append(svm2.score(x_test, y_test))

# Plotting
plt.figure(figsize=(15, 5))
sns.lineplot(x=range(1, 200), y=train_score_list, label='Train Score', marker='o')
sns.lineplot(x=range(1, 200), y=test_score_list, label='Test Score', marker='*')

plt.xlabel('C value')
plt.ylabel('Accuracy')
plt.title('SVM Accuracy vs. C value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x_data)


svm = SVC(kernel='linear')  
svm.fit(X_pca, y_data)


x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 500),
    np.linspace(y_min, y_max, 500)
)


Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm', edgecolor='k', s=60)
plt.title("SVM Decision Region after PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()
```
